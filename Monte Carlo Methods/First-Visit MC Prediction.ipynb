```python
def isTerminal(state):
    if state<=0 or state == len(value_list)-1:
        return True
    return False
```

```python
def get_action():
    return np.random.choice(action_space)
```

```python
def step(current_state, action):
    action = action
    next_state = int(current_state) + int(action)
    reward = rewards[int(next_state)]
    return next_state, reward
```

```python
def calculate_reward(index):
    g = rewards_list[index]
    for reward in rewards_list[index+1:]:
        g = gamma * g + reward
    return g
```

```python
def calculate():
    visitedMap=[]
    for i in isVisited[:-1]:
        print(i)
        if i not in visitedMap:
            visitedMap.append(i)
            value_list[i] = calculate_reward(isVisited.index(i))
        else: 
            continue
    return value_list
```
